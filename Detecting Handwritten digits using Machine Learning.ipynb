```python
from sklearn.datasets import load_digits
%matplotlib inline
import matplotlib.pyplot as plt
handwrit_digits = load_digits()
```

```python
plt.gray()
for i in range(5):
    plt.matshow(handwrit_digits.images[i]) 
```

```python
dir(handwrit_digits)
```

```python
handwrit_digits.data[0]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(handwrit_digits.data,handwrit_digits.target, test_size=0.2)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(handwrit_digits.data[0:5])
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_predicted)
confusion_matrix
```

```python
import seaborn as sn
plt.figure(figsize = (12,8))
sn.heatmap(confusion_matrix, annot=True)
plt.xlabel('Prediction')
plt.ylabel('Actual')
```
